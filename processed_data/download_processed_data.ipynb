```python
def process_n_save(source, temp_dir, save_dir, year):
    
    year = str(year)
    
    shutil.unpack_archive(source, os.path.join(temp_dir, f'data_{year}'))
    
    files = os.listdir(os.path.join(temp_dir, f'data_{year}', f"/NW_weather_models_2D_parameters_{year}"))
    
    for file in files:
        os.mkdir(os.path.join(temp_dir, 'Extracted', file))
        shutil.unpack_archive(os.path.join(temp_dir, f'data_{year}', f"/NW_weather_models_2D_parameters_{year}", file),os.path.join(temp_dir, 'Extracted', file))
    
    # Save Wind
    text_files = glob.glob(str(os.path.join(temp_dir, 'Extracted')) + "/**/arome_10m_NW_*.grib", recursive = True)
    for file_ in text_files:
        shutil.copy(file_, os.path.join(save_dir, 'Wind'))
```
