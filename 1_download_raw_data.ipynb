```python
import numpy as np
import pandas as pd 
import os
import shutil
import glob
```

```python
dataset_path = './dataset'
temp_dir = './temp'

os.makedirs(dataset_path, exist_ok=True)
os.makedirs(temp_dir, exist_ok=True)
```

```python
meteonet_url_form = 'https://meteonet.umr-cnrm.fr/dataset/data/NW/weather_models/2D_parameters/NW_weather_models_2D_parameters_{year}.tar.gz'

output_file_name_form = os.path.join(temp_dir, 'meteonet_{year}.tar.gz')
```

```python
def process_n_save(source, temp_dir, save_dir, year):
    
    year = str(year)
    
    shutil.unpack_archive(source, os.path.join(temp_dir, f'data_{year}'))
    
    files = os.listdir(os.path.join(temp_dir, f'data_{year}', f"/NW_weather_models_2D_parameters_{year}"))
    
    for file in files:
        os.mkdir(os.path.join(temp_dir, 'Extracted', file))
        shutil.unpack_archive(os.path.join(temp_dir, f'data_{year}', f"/NW_weather_models_2D_parameters_{year}", file),os.path.join(temp_dir, 'Extracted', file))
    
    # Save Wind
    text_files = glob.glob(str(os.path.join(temp_dir, 'Extracted')) + "/**/arome_10m_NW_*.grib", recursive = True)
    for file_ in text_files:
        shutil.copy(file_, os.path.join(save_dir, 'Wind'))
```

# 1. Download from Meteonet Data

## 1.1 2016 Data

```python
year = 2016

print(meteonet_url_form.format(year=year))
print(output_file_name_form.format(year=year))
```

```python
url = meteonet_url_form.format(year=year)
file_down_path = output_file_name_form.format(year=year)

```

```python
!wget url $url -O $file_down_path
```

```python

process_n_save(file_down_path, temp_dir, dataset_path, year)
```
